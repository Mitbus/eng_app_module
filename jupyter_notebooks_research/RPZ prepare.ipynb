```python
from matplotlib import pyplot as plt
def print_part_plot(model, top_words):
    in_model = 0
    total = 0
    x = []
    y = []
    res = []
    for w in top_words:
        if w in model.wv:
            in_model += 1
        else:
            res.append(w)
        total += 1
        x.append(total)
        accuracy = in_model / total
        y.append(accuracy)
    plt.plot(x, y)
    return res
```
