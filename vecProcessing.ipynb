```python
import io
```

```python
def load_vectors(fname):
    fin = io.open(fname, 'r', encoding='utf-8', newline='\n', errors='ignore')
    n, d = map(int, fin.readline().split())
    data = {}
    k = 0
    for line in fin:
        k += 1
        if k > 20:
            break
        tokens = line.rstrip().split(' ')
        data[tokens[0]] = map(float, tokens[1:])
    return data
```

```python
fname = 'wiki.en.vec'
```

```python
load_vectors(fname)
```
